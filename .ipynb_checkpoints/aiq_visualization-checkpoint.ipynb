```python
import pandas as  pd
import matplotlib.pyplot as plt
from config import DB_CONFIG
from sqlalchemy import create_engine
engine = create_engine(f"mysql+mysqlconnector://{DB_CONFIG['user']}:{DB_CONFIG['password']}@{'localhost'}/{DB_CONFIG['database']}")
```

```python
# merged_sale_user__weather_data=pd.read_csv("/content/drive/MyDrive/aiq/adata.csv")
merged_sale_user__weather_data= pd.read_sql("""SELECT  st.*,t1.*,t2.*
FROM  user_Location_info  as t1 RIGHT join user_info as t2
on t1.user_info_table_id= t2.id
join Sales_info  st 
on st.customer_id=t2.id""", con=engine)
```

```python
# Calculate total sales amount per customer
total_sales_per_customer = merged_sale_user__weather_data.groupby('name')['price'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(total_sales_per_customer['name'], total_sales_per_customer['price'])
plt.xlabel('Customer')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount per Customer')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Determine the average order quantity per product
average_order_quantity_per_product = merged_sale_user__weather_data.groupby('product_id')['quantity'].mean().reset_index()
average_order_quantity_per_product.head()
```

```python
# Identify the top-selling products
top_selling_products = merged_sale_user__weather_data.groupby('product_id')['quantity'].sum().nlargest(5)
# top_selling_products.head()
plt.figure(figsize=(10, 6))
top_selling_products.plot(kind='bar', color='skyblue')
plt.xlabel('Product ID')
plt.ylabel('Total Quantity Sold')
plt.title('Top Selling Products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Identify the top-selling products or customers
top_selling_customers = merged_sale_user__weather_data.groupby('name')['price'].sum().nlargest(5)
# top_selling_customers.head()
plt.figure(figsize=(10, 6))
top_selling_customers.plot(kind='bar', color='skyblue')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales Amount')
plt.title('Top Selling Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyze sales trends over time
merged_sale_user__weather_data['order_date'] = pd.to_datetime(merged_sale_user__weather_data['order_date'])
sales_trends_over_time = merged_sale_user__weather_data.groupby(merged_sale_user__weather_data['order_date'].dt.to_period('M'))['price'].sum()
# sales_trends_over_time.head()
plt.figure(figsize=(10, 6))
plt.plot(sales_trends_over_time.index.to_timestamp(), sales_trends_over_time.values, marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.title('Sales Trends Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate average sales amount per weather condition
average_sales_per_weather = merged_sale_user__weather_data.groupby('weather_main')['price'].mean().reset_index()
average_sales_per_weather.head()
plt.figure(figsize=(10, 6))
plt.bar(average_sales_per_weather['weather_main'], average_sales_per_weather['price'], color='skyblue')
plt.xlabel('Weather Condition')
plt.ylabel('Average Sales Amount')
plt.title('Average Sales Amount per Weather Condition')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate total sales amount per customer
total_sales_per_customer = merged_sale_user__weather_data.groupby('address.city')['price'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(total_sales_per_customer['address.city'], total_sales_per_customer['price'])
plt.xlabel('address.city')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount per City')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
